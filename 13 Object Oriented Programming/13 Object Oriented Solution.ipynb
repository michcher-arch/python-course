```python
#Q1
class Clock:
  max_time = 0
  
  def __init__(self,mins,hours) :
    self._secs = 0
    self._mins = mins
    self._hours = hours
    Clock.max_time = max(Clock.max_time, hours)

  def set_secs(self,secs) :
    self._secs = secs
    
  def set_mins(self,mins) :
    self._mins = mins

  def set_hours(self,hours) :
    self._hours = hours

  def get_secs(self) :
    return self._secs
    
  def get_mins(self) :
    return self._mins

  def get_hours(self) :
    return self._hours

  secs = property(get_secs,set_secs)
  mins = property(get_mins,set_mins)
  hours = property(get_hours,set_hours)

  def tick(self) :
    self._secs += 1
    if(self._secs == 60) :
      self._secs = 0
      self._mins += 1
      if(self._mins == 60) :
        self._mins = self._mins = 0
        self._hours = (self._hours + 1) % 24
    Clock.max_time = max(Clock.max_time,self._hours)

  def __str__(self) -> str:
    return  f'{self._hours}:{self._mins}:{self._secs}'

c = Clock(1,1)
print(c)
for i in range(3600) :
  c.tick()
print(c)
print(Clock.max_time)
```

```python
#Q2
class Book :
  def __init__(self,name,copies,cost) :
    self._name = name
    self._copies = copies
    self._cost = cost

  def get_name(self):
    return self._name
  def get_copies(self):
    return self._copies
  def get_cost(self):
    return self._cost
  def set_name(self,name) :
    self._name = name
  def set_copies(self,copies) :
    self._copies = copies
  def set_cost(self,cost) :
    self._cost = cost

  name = property(get_name,set_name)
  copies = property(get_copies,set_copies)
  cost = property(get_cost,set_cost)

  def __str__(self) -> str:
    return f'name: {self._name} copies: {self._copies} cost: {self._cost}'
  
  def cost(self) :
    return self._cost * self._copies
```
