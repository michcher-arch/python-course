```python
# Q2
import numpy as np

def shorashim(a,b,c):
  sul1 = 0
  sul2 = 0

  sul1 = b*-1
  sul2 = b*-1
  if(pow(b,2)-(4*a*c) > 0):
    sul1 += np.sqrt(pow(b,2)-(4*a*c))
    sul2 -= np.sqrt(pow(b,2)-(4*a*c))
    sul1 = sul1/(2*a)
    sul2 = sul2/(2*a)
  else:
      return 

  return np.array([sul1,sul2])
```

```python
# Q4
import numpy as np

def above_avg(arr) :
  avg = np.average(arr)
  print('avg', avg)
  filter_arr = arr >= avg
  arr2 = arr[filter_arr]

  return arr2

above_avg(np.array([1, 2, 3, 1, 2, 3]))
```

```python
# Q5

def is_perfect(num) :
  sum = 0
  for i in range(1,int(num/2) + 1) :
    if(num % i ==0) :
      sum+=i
  return sum == num
```

```python
# Q6

def all_perfects() :
  for i in range(1, 1001) :
    if(is_perfect(i)) :
      print(i)
```

```python
# Q7

def ribuim(begin = 1, end = 30) :
  arr3 = np.arange(end - begin + 1)
  j = 0
  for i in range (begin, end + 1) :
    arr3[j] = i * i
    j+=1
  return arr3
```

```python
# Q8

def factorial(n) :
  if(n==0) :
    return 1
  return factorial(n - 1) * n
```
