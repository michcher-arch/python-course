```python
# Q2
val = input('What is your salary?\n')
amount = int(val)
tax = 0
if amount > 5000 :
  tax = tax + 5000 * 0.1
  print('salary is above 5000, adding 500 for the first 5000')
  amount = amount - 5000
  if amount > 3000 :
    tax = tax + 3000 * 0.15
    print('salary is above 8000, adding 450')
    amount = amount - 3000
    if amount > 4000 :
      tax = tax + 4000 * 0.22
      print('salary is above 12000, adding 880')
      amount = amount - 4000
      if amount > 8000 :
        tax = tax + 8000 * 0.3
        print('salary is above 20000, adding 2400')
        amount = amount - 8000
        step = amount * 0.5
        tax = tax + step
        print('Adding that rest. that is', step)
      else :
        step = amount * 0.3
        tax = tax + step
        print('salary is less than 20000, adding', step)
    else :
      step = amount * 0.22
      tax = tax + step
      print('salary is less than 12000, adding', step)
  else :
    step = amount * 0.15
    tax = tax + step
    print('salary is less than 8000, adding', step)
else :
  tax = tax + amount * 0.1
  print('salary is less than 5000, adding', tax)

print('Total tax', tax)








```
