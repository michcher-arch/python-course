```python
#Q1
def secret(st, dic):
  ret = ""
  for ch in st:
    if ch in dic:
      ret += dic[ch]
    else:
      ret += ch
  return ret;

dic1 = {
    "q" : "i",
    "w" : "l",
    "e" : "o",
    "r" : "v",
    "t" : "e",
    "y" : "p",
    "u" : "y",
    "i" : "t",
    "o" : "h",
    "a" : "n"
}

print(secret("q wert yuioea", dic1))
```

```python
def func2(grades):
  passed = {}
  failed = {}
  for student in grades:
    if grades[student] > 54:
      passed[student] = grades[student]
    else:
      failed[student] = grades[student]

  for student in failed:
    failed[student] *= 1.2

  total = {}
  total.update(passed)
  total.update(failed)

  list1 = list(total.values())
  list1.sort()

  list_items = list(total.items())
  max_key = list_items[0][0]
  min_key = list_items[0][0]

  for i in range(len(list_items)) :
    if list_items[i][1] > total[max_key] :
      max_key = list_items[i][0]
    if list_items[i][1] < total[max_key] :
      min_key = list_items[i][0]

  del total[max_key]
  del total[min_key]

  return total

grades = {
  "a" : 100,
  "b" : 54,
  "c" : 60,
  "d" : 30,
  "e" : 25,
  "f" : 90
}

print(func2(grades))
```

```python
#Q3
def count_letters(st):
  dic = {}
  for ch in st:
    if ch != " ":
      if ch in dic:
        dic[ch] += 1
      else:
        dic[ch] = 1

  keys = dic.keys()
  list1 = list(keys)  
  list1.sort()
  for key in list1 :
    print(key,':', dic[key])

st = "aaa bb fgcc zza"
count_letters(st)
```

```python
#Q4
def unique(dic):
  ret = {}
  for item in dic:
    ret[dic[item]] = 1
  for item in ret:
    print(item)

dic = {
    1 : "a",
    2 : "b",
    3 : "a",
    4 : "c",
    5 : "b"
}

unique(dic)
```
