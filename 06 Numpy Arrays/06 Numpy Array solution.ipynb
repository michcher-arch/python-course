```python
import numpy as np

#1
LENGTH = 15

#arr = np.empty(LENGTH, dtype='i')
arr = np.array([10, 20, 30, 40, 100])
print('before user input', arr)

#for i in range(LENGTH) :
#  value = input('please enter a grade')
#  num = int(value)
#  arr[i] = num

print('after user input', arr)
print(arr)

#2
arr_copy = arr.copy()
print('arr_copy', arr_copy)

#3
arr_copy.sort()
print('arr_copy', arr_copy)

#4
arr_low = arr_copy[0:3]
arr_high = arr_copy[-3:]
print('arr_low', arr_low)
print('arr_high', arr_high)

#5
arr_split = np.array_split(arr_copy, 3)
print('arr_split', arr_split)
print('arr_split type', type(arr_split))
print(arr_split[0])

# 6
grade_100 = np.where(arr_copy == 100)
print('grade 100', grade_100)
print(grade_100[0].size)

#7 
filter_arr = arr_copy < 55
failed_arr = arr_copy[filter_arr]
print('failed_arr', failed_arr)

print('failed arr avg', np.average(failed_arr))

#8
failed_arr = failed_arr * 120 // 100
print('failed arr after bonus', failed_arr)

#9
filter_arr = arr_copy > 54
pass_arr = arr_copy[filter_arr]
print('pass_arr', pass_arr)

#10 
final_arr = np.concatenate((failed_arr, pass_arr))
print('final_arr', final_arr)
```
