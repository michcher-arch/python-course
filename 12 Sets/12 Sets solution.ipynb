```python
#Q1
def remove_dups(list1):
  return list(set(list1))

list1 = [3, 3, 4, 5, 6, 7]
remove_dups(list1)
```

```python
#Q2
def remove_string_dups(str1) :
  list1 = list(str1)
  list2 = remove_dups(list1)
  str2 = str(list2)

  return str2

print(remove_string_dups('aaabbbbccc'))
```

```python
#Q3 
def two_lists(list1, list2) :
  list3 = list1 + list2
  return remove_dups(list3)

print(two_lists([1, 1, 1, 2, 2, 2, 3, 3, ], [1, 1, 1, 2, 3, 3]))
```

```python
#Q4
def isSubSet(set1, set2):
  if (set1 < set2):
    print("sub set")
  if set1 == set2:
    print("same")
  

isSubSet({1, 2}, {1, 2, 3})
isSubSet({1, 2, 3}, {1, 2, 3})
```

```python
#5
def removeFromSet(set2, set1):
  for x in set1:
    if (x in set2): set2.remove(x)
  return set2

removeFromSet({1, 2, 3, 4}, {2, 4})
```
