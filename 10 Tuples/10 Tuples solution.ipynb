```python
# Q1
def prime(min=1, max=100):
    list_prime = []
    dividable = False
    for i in range(min, max + 1):
        for j in range(1, i):
            if i % j == 0 and j != 1:
                dividable = True
                break
        if not dividable:
            list_prime.append(i)
        dividable = False
    if len(list_prime) < 3:
        print("not enough numbers found")
        return None
    else:
        tuple_prime = tuple(list_prime[:4])
    return tuple_prime
```

```python
# Q2
def func1():
    tuple_func1 = prime()
    num1, num2, num3, num4 = tuple_func1
    print(num1)
    print(num2)
    print(num3)
    print(num4)
```
