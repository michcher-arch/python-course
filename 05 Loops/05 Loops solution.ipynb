```python
# Q8 do-while

sum = 0
count = 0

while(True) :
  n = int(input('Enter grade '))
  if n < 1 or n > 100 :
    print('Not a valid grade, please enter again')
    continue

  count += 1
  sum += n
  avg = sum//count

  if avg > 70:
      break

print('avg', avg)
```
