```python
# Q2
def targil2(names):
    list2 = names.split(',')
    list2.sort()
    current_start = (list2[0])[0]
    for name in list2:
        if name[0] != current_start:
            print("\n")
        print(name)
        current_start = name[0]
```

```python
# Q3
def targil3(fruits):
    list3 = fruits.split()
    list3.sort(key=len)
    for fruit in list3:
        print(fruit + "\n")
```
