```python
# Q1
def q1(s1, s2) :
  if len(s1) != len(s2) :
    print(s1, s2, 'not in the same length!')
    return 

  s = ''
  for i in range(len(s1)) :
    s = s + s1[i] + s2[i]

  return s
```

```python
# Q2
def q2(s, n) :
  if n >= len(s) or n < 0 :
    print(n, 'Illigal value for n')
    return

  s1 = s[:n]
  s2 = s[n+1:]

  return s1 + s2
```

```python
# Q3
def change(s, c = '*') :
  first = s[0]
  rest = s[1:]
  x = rest.replace(first, c)
  
  return first + x
```

```python
# Q5 
import numpy as np

def sort_string(str) :
  arr = convert_string_to_array(str)

  arr2 = np.sort(arr)
  str2 = convert_arr_to_string(arr2)

  return str2

def convert_string_to_array(str) :
  arr = np.empty(len(str), dtype='U')
  for i in range(len(str)) :
    arr[i] = str[i]

  return arr

def convert_arr_to_string(arr) :
  str = ''
  for c in arr :
    str = str + c

  return str
```

```python
# Q6
import numpy as np

def words(str) :
  arr = split_string_to_array(str)
  arr2 = np.sort(arr)
  prev = ''
  for word in arr2 :
    if word != prev :
      print(word)
      prev = word


def split_string_to_array(str) :
  n = str.count(',') + 1
  arr = np.empty(n, dtype = 'O')
  word_count = 0
  word = ''
  for c in str :
    if c == ',' :
      arr[word_count] = word
      word = ''
      word_count = word_count + 1
    else :
      word = word + c

  arr[-1] = word
  return arr
```
